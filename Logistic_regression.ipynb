```python
#libraries required for the logistic regression
import pandas as pd #to read a data from different resource
import matplotlib.pyplot as plt #visiualization purpose
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("iris.csv")
df.head()
```

```python
df['variety'].unique()
```

```python
df=df[df['variety']!="Setosa"]
```

```python
df.head()
```

```python
#to check weather any null values
df.isnull().sum()
```

```python
df['variety'].value_counts()
```

```python
#converting the output into binary
df['variety'] = df['variety'].map({'Versicolor': 0, 'Virginica': 1})
df.head()
```

```python
#spliting the data into dependent and independent
X=df.iloc[:,:-1] #it will select all row and columns execpt last one column
y=df.iloc[:,-1]#This operation effectively selects the last column of the DataFrame
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_reg=GridSearchCV(log_reg,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_reg.fit(X_train,y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
print(classifier_reg.best_score_)
```

```python
#predication
y_predic=classifier_reg.predict(X_test)
```

```python
#accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_predic,y_test)
print(score)
```

```python
print(classification_report(y_predic,y_test))
```
